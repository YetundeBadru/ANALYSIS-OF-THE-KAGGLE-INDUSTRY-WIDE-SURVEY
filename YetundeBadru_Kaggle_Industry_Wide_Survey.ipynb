### Read the raw data into a dataframe

```python
# import the libraries
import pandas as pd 
import numpy as np 
```

```python
# Read the data into a dataframe 
survey = pd.read_csv('2019multiple_choice_responses.csv', low_memory=False)
```

```python
# Read the first 5 rows
survey.head(5)
```

---

### Subsetting the 1st 11 columns

```python
# This gives a view of the columns title
survey.columns[:11]
```

```python
# Subset the first 11 columns
col_subset = survey[['Time from Start to Finish (seconds)', 'Q1', 'Q2', 'Q2_OTHER_TEXT', 
                     'Q3', 'Q4', 'Q5', 'Q5_OTHER_TEXT', 'Q6', 'Q7', 'Q8']]
```

```python
# viewing the subsetted columns
col_subset.head()
```

---

### Renaming the 11 columns

```python
# After viewing the subsetted dataframe and columns, its obvious that our data looks raw and messy
# I'll have to rename the columns for easier EDA
```

```python
# read the first row that has the descriptions of the column names
# This helps to understand the columns and properly rename it.
col_subset.head(1).values         
```

```python
# Renaming the columns using key & value pairs
col_subset = col_subset.rename(columns={'Time from Start to Finish (seconds)':'Duration(secs)',
                       'Q1':'Age', 
                       'Q2':'Gender', 
                       'Q2_OTHER_TEXT':'Gender_(Other)', 
                       'Q3':'Country', 
                       'Q4':'Education', 
                       'Q5':'Job_title', 
                       'Q5_OTHER_TEXT':'Job_title(Other)', 
                       'Q6':'company_size', 
                       'Q7':'dataScience_team',
                       'Q8':'ML_adoption'
                      }
                 )
```

```python
# drop the first row that has the descriptions of the column names 

col_subset.drop(index=0, inplace=True)
```

```python
col_subset.head()
```

```python
col_subset.columns
```

```python
## REASON BEHIND RENAMING THE COLUMNS
```

### The columns were renamed using clear and brief labels based on the entry(values) provided in the first row of the dataset, they offer a clear explanation of the values contained in each column.

##### 1. 'Duration(secs)' - It is the amount of time for completing the survey
##### 2. 'Age' - Age of the respective respondents
##### 3. 'Gender' - Gender of the respective respondents 
##### 4. 'gender_other' - Alternative gender category, if any.
##### 5. 'country' - Country of residence of the respective respondents.
##### 6. 'education_level' - The educational level of the respective respondents
##### 7. 'job_title' - The current job role of each respondent
##### 8. 'job_title_other' - An alternative job role, if any.
##### 9. 'employer_size' - The size of the respondent's company/employer
##### 10. 'employee_experience' - The employee's experience using the employees workload
##### 11. 'ml_adoption' - Indicating the respondent's current company/employer adopts Machine Learning in their business

### Share at least two insights about the educational levels of at least two countries (from different continents) using visuals.

```python
# import the libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

### France                       (Europe)

```python
# create new subset for France's data only 

france = subset[subset['Country'] == 'France']
france.head()
```

```python
# count plot using seaborn for visualization
sns.countplot(x='Country', hue='Education', stat='count', data=france)
plt.legend(bbox_to_anchor=(1, 0.85), loc='upper right', fontsize='small')
plt.title("France Educational Level")
plt.show()
```

### Insights from France

* Majority of the individuals have a Masters's degree which suggests that france has a very good education system and that gave room for individual advance education.
* There is almost 100 number of individuals that pursued doctoral advanced degree, this explains that most individuals get jobs with their undergraduate degree.
* France has a very low number of individuals that has no formal education or individuals who went to college without bachelor degree.

---

### India                        (Asia)

```python
# create new subset for India's data only 

india = subset[subset['Country'] == 'India']
india.head()
```

```python
# count plot using seaborn for visualization
sns.countplot(x='Country', hue='Education', stat='count', data=india)
plt.legend(bbox_to_anchor=(1, 0.85), loc='upper right', fontsize='small')
plt.title("India's Educational Level")
plt.show()
```

### Insights from India

* Majority of the individuals have a bachelor's degree which explains that undergraduate education is very accessible in India.
* There is over 1,500 number of individuals that pursued master's degree, this explains that advanced study is also accessible .
* India has a very low number of individuals that with no formal education, and individuals who went to college without bachelor degree.
* Less than 500 individuals have advanced doctoral degree, this explains the high job opportunities for individual with bachelor and master's degree, so there really is no need for doctoral advancement.

### South Africa                 (Africa)
